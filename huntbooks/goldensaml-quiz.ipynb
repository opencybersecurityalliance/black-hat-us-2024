# GoldenSAML Quiz

#### Dataset

The same 4 datasources as in our main GoldenSAML huntbook.

The datasources will be automatically listed in the `FROM` clause of a `GET` command when pressing `tab` to auto-complete.

```kestrel
# the initial step of the quiz huntbook

all_events = GET event FROM sqlalchemy://GoldenSAML-WindowsEvents
             WHERE device.hostname LIKE '%'
             START 2021-08-02T00:00:00.000Z STOP 2021-08-03T00:00:00.000Z
             
DISP all_events ATTR time, type_uid, type_name, device.hostname SORT BY time ASC
```

## Quiz A: Find Logged-in Users

1. filter logon events into a new Kestrel variable
2. use `FIND` command to extract users `RESPONDED` to the events
3. display username (`name`), SID (`uid`), and logon hostname (`endpoint.hostname`) using `DISP` command 

## Quiz B: Find Accessed Objects in WindowsEvents Log

1. filter object operation events into a new Kestrel variable
2. use `FIND` command to extract objects (`managed_entity` type in OCSF) which `RESPONDED` to the events
3. show the details of the objects

## Quiz C: Object Access Details in Microsoft365DefenderEvents Log

**Prerequisite: quiz B**

The `WindowsEvents` has very limited information on the accessed objects. Fortunately, the objects are also logged by MS Defender. Let's check the `Microsoft365DefenderEvents` log source to find more.

1. plan to search for `query_info` in `Microsoft365DefenderEvents` log source containing the id (`uid`) of the object from Quiz B
2. read the OCSF doc on `query_info` object to check its fields
3. pick up one or more fields to write the Kestrel `WHERE` clause (you can `OR` multiple comparison expressions in the `WHERE` clause)
4. copy the `uid` from Quiz B to form the `WHERE` clause here
5. use `LIKE` and `%...%` to maximize the possibility of a match
6. write your `GET` command followed by a `DISP`
